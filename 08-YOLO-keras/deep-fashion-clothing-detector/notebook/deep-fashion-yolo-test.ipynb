```python
# -*- coding: utf-8 -*-

%matplotlib inline

import random

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import numpy as np
```

```python
DATA_DIR = '/Users/admin/Documents/deep_fashion/fashion_landmark_detection'
```

```python
x1, y1, x2, y2 = 53, 130, 289, 440

img = plt.imread(DATA_DIR + '/Img/img/img_00000001.jpg')
fig, ax = plt.subplots()
ax.imshow(img)
img_width = img.shape[1]
img_height = img.shape[0]

width = x2-x1
height = y2-y1
x1 = x1 - round(((random.random()-0.5) * 0.2) * width)
y1 = y1 - round(((random.random()-0.5) * 0.2) * height)

if x1 < 0: x1 = 0
if x1+width >= img_width: width -= (x1 + width - img_width)
if y1 < 0: y1 = 0
if y1+height >= img_height: height -= (y1 + height - img_height)

rect = patches.Rectangle((x1, y1), width, height,linewidth=1,edgecolor='r',facecolor='none')
ax.add_patch(rect)
plt.show()
print(x1, y1, y1+height, x1+width)
print(img.shape)
```
